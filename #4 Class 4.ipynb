```python
import math
def candy(num, a=25, b=17, c=12):
    n1 = math.floor(num/a)
    n2 = math.floor(num/b)
    n3 = math.floor(num/c)
    if num%a == 0:
        return "Yes: " + str(n1) + " of " +str(a)
    elif num%b == 0:
        return "Yes: " + str(n2) + " of " +str(b)
    elif num%c == 0:
        return "Yes: " + str(n3) + " of " +str(c)
    
    for i in range(n1+1):
        for j in range(n2+1):
            if (num - i*a - j*b)%c == 0 and (num - i*a - j*b) >= 0:
                n3 = math.floor((num - i*a - j*b)/c)
                return "Yes: " + str(i) + " of " +str(a) + ", " + str(j) + " of " + str(b) + ", " + str(n3) + " of " +str(c)
    
    return "No"
```

```python
ret = candy(196, 12, 15, 35)
print(ret)
```

```python
def prod(a, b):
    return a*b
def prod2(a, b):
    if a == 1:
        return b
    return prod(a-1, b) + 5
def fib(n):
    if n == 1 or n==2:
        return 1
    return fib(n-2) + fib(n-1)

def fiboptim(n, store=dict()):
    if n == 1 or n==2:
        store[n] = 1
        return 1
    
    if n-1 in store.keys():
        v1 = store[n-1]
    else:
        v1 = fiboptim(n-1,store)
        store[n-1] = v1
    if n-2 in store.keys():
        v2 = store[n-2]
    else:
        v2 = fiboptim(n-2,store)
        store[n-2] = v2
    v = v1 + v2
    store[n] = v
    return v
```

```python
import time

start = time.time()
fb = fiboptim(30000)
stop = time.time()
df = stop -start
print("fib: " + str(fb))
print(df)
```
