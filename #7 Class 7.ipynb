```python
def getPie(N):
    n = 0;
    r = 1
    Xc = []
    Xs = []
    Yc = []
    Ys = []
    for i in range(N):
        pt = np.random.rand(2)
        x = 2*pt[0] - 1
        y = 2*pt[1] - 1
        if x**2 + y**2 <= r**2:
            n += 1
            Xc.append(x)
            Yc.append(y)
        else:
            Xs.append(x)
            Ys.append(y)
    p = 4*n/N
    fig = plt.figure()
    plt.plot(Xc,Yc, "ro")
    plt.plot(Xs, Ys, "bo")
    return p
```

```python
getPie(1000)
```
