### 3D Plots

#### Line 3D Plot

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
x = np.linspace(0,100,20)
y = x;
z = np.sqrt(y + x)
# print(z)
%matplotlib inline

fig = plt.figure()
ax = fig.gca(projection='3d') # Get Current Axes
ax.plot(x,y,z, "ro-", lw=4)
plt.xlabel("X Values")
plt.ylabel("Y Values")
ax.set_zlabel("Z Values")
ax.view_init(azim=18,elev=10)
plt.show()



```

#### Surface Plot

```python
rg = np.linspace(-5,5,100) # Define the range along X and Y axis
# print(rg)
X,Y = np.meshgrid(rg,rg)  # Define the meshgrid. This produces a matrix
# print(X,Y)
Z = -20*np.exp(-0.2*np.sqrt(0.5*(X**2 + Y**2))) - np.exp(0.5*(np.cos(2*np.pi*X) + np.cos(2*np.pi*Y))) + np.exp(1) + 20  # This is the equation for Z
fig = plt.figure() # Create the figure for plotting
ax = fig.gca(projection="3d") # Create axes. The axes is inside the figure
ax.plot_surface(X,Y,Z,cmap=cm.jet, linewidth=1, antialiased=True) # Do the plotting
plt.show() # Show the plot


```

```python
rg = np.linspace(-5,5,100) # Define the range along X and Y axis
# print(rg)
X,Y = np.meshgrid(rg,rg)  # Define the meshgrid. This produces a matrix
# print(X,Y)
Z = -(X**2 + Y**2) + 4 # This is the equation for Z
fig = plt.figure() # Create the figure for plotting
ax = fig.gca(projection="3d") # Create axes. The axes is inside the figure
ax.plot_surface(X,Y,Z,cmap=cm.jet, linewidth=1, antialiased=True) # Do the plotting
plt.show() # Show the plot
```

### Patches

```python
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
# Triangle
fig = plt.figure()
ax = fig.gca()
points = [ [1,1], [10,1], [5,7] ]
points2 = [ [10,1], [20,1],[25,7], [15,7] ]
trig = Polygon(np.array(points), True)
trig2 = Polygon(np.array(points2), True)
patches = [trig, trig2]
collection = PatchCollection(patches, cmap=cm.plasma)
colors = 100*np.random.rand(len(patches))
print(colors)
collection.set_array(np.array(colors))
ax.add_collection(collection)
ax.set_xlim(0,27)  # Set the limit of the x axis
ax.set_ylim(0,10)  # Set the limit of the y axis
plt.show()

    
```

```python
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
# Triangle
fig = plt.figure()
ax = fig.gca()
points = [ [5,8], [2,6], [3,4],[5,5],[4,1],[11,1],[10,5],[12,4],[14,6],[10,8],[9,6.5], [6,6.5] ]
trig = Polygon(np.array(points), True)
patches = [trig]
collection = PatchCollection(patches, cmap=cm.plasma)
colors = 100*np.random.rand(len(patches))
collection.set_array(np.array(colors))
ax.add_collection(collection)
ax.set_xlim(0,15)  # Set the limit of the x axis
ax.set_ylim(0,8)  # Set the limit of the y axis
plt.show()

    
```
